```python
import pandas as pd
df = pd.read_csv('recent-grads.csv')
df.shape
```

```python
df.head()
```

```python
#Method 1: Boolean mask
df1 = df[df['Major_category'] == 'Social Science']
print(df1.shape)
df1.head()
```

```python
#alternate process
filter_criteria = (df['Major_category'] == 'Social Science') & (df['ShareWomen'] >= .5)
df1m = df[ filter_criteria ]

print(df1m.shape)
df1m.head()
```

```python
#to get the opposite of our filter criteria using ~
df1_op = df[~filter_criteria]
print(df1_op.shape)
df1_op.head()
```

```python
#Method 2 :loc
df2 = df.loc[filter_criteria]
print(df2.shape)
df2.head()
```

```python
df2 = df.loc[filter_criteria, 'Part_time']
df2.head()
```

```python
cols = ['Part_time', 'Sample_size', 'Total']
df2 = df.loc[filter_criteria, cols]
df2.head()
```

```python
#Method 3: Query
df3 = df.query("Major_category =='Social Science' & ShareWomen >= 0.5")
print(df3.shape)
df3.head()
```

```python
#Method 4: selecting rows and columns
df4 = df.iloc[:10, 2:6]
print(df4.shape)
df4.head()
```

```python
#Method 5: loc
df5 = df.loc[df.index[:10], ['Major', "Major_category"]]
print(df5.shape)
df5.head()
```

```python
df5 = df.set_index(df['Major'])
df5.head()
```

```python
rows = ["PETROLEUM ENGINEERING", "MECHANICAL ENGINEERING"]
cols = ['Major', 'Major_category']
df5 = df5.loc[rows, cols]
print(df5.shape)
df5.head()
```

```python
#Method 6: filtering out missing data using notnull()
df6 = df.copy()
df6.isnull().sum()
```
